```python
import pandas as pd
import googleapiclient.discovery as api
import json
import time
```

```python
API_KEY = "" # your API_KEY here. Get it from https://console.cloud.google.com
api_version = "v3"
api_service_name = "youtube"

youtube = api.build(api_service_name,
    api_version,
    developerKey = API_KEY)
```

```python
# At first let's take videos and create video DataFrame
```

```python
part = "snippet"
q = "импортозамещение" # put here your q you look for
type = "video"
amount = 50
```

```python
request = youtube.search().list(
    part = part,
    maxResults = amount,
    q = q,
    type = type
)
response = request.execute()

df_supplemented = pd.json_normalize(response['items'])
df_supplemented
```

```python
df_supplemented['snippet.publishedAt'] = df_supplemented['snippet.publishedAt'].apply(lambda yyyy: yyyy[:4])
df_supplemented['snippet.publishedAt']
```

```python
# okay next step is important part
# we will take all videos that we possibly can get
# your API_KEY may run out of quotas
# if so comment out first line in code change your key and go again
```

```python
year = int(min(df_supplemented['snippet.publishedAt'])) # comment out this line if you have no quotas left
i = 0 
while year < 2024:
    request = youtube.search().list(
        part = part,
        maxResults = amount,
        q = q,
        type = type,
        publishedAfter = f'{year}-01-01T00:00:00Z'
    )
    response = request.execute()
    
    df_add = pd.json_normalize(response['items'])
    df_supplemented = pd.concat([df_supplemented, df_add])
    
    while 'nextPageToken' in response.keys():
        request = youtube.search().list(
            part = part,
            maxResults = amount,
            q = q,
            type = type,
            pageToken = response['nextPageToken'],
            publishedAfter = f'{year}-01-01T00:00:00Z'
        )
        response = request.execute()
        
        print(f'iteration №{i}, "date" {year}')
        i += 1
        
        df_add = pd.json_normalize(response['items'])
        df_supplemented = pd.concat([df_supplemented, df_add])
        
    print(f"Everything from {year} {response['pageInfo']['totalResults']}, \
          and we found: {len(df_supplemented.drop_duplicates(f'id.{type}Id'))}")
    
    year += 1
```

```python
video_df = df_supplemented.drop_duplicates(f'id.{type}Id')
video_df.index = range (1, len (video_df) + 1) 
display(video_df)
```

```python
# now to look for video let's take the channel id
videoId_list = video_df[f'id.{type}Id'].to_list()
videoId_list
```

```python
request = youtube.videos().list(
    part = ["snippet",
            "id",
            "topicDetails",
            "statistics", 
            "status"],
    id = videoId_list[0:50], 
    maxResults = amount)

response = request.execute()

df_supplemented = pd.json_normalize(response['items'])
df_supplemented
```

```python
i = 0
for i in range(50, len(videoId_list)+1, 50):
    request = youtube.videos().list(
       part = ["snippet",
            "contentDetails",
            "localizations", 
            "statistics", 
            "status", 
            "topicDetails"], 
    id = videoId_list[i:i+50], 
    maxResults = 50)
    response = request.execute()
    
    print(f'iteration №{i}')
    
    df_additional = pd.json_normalize(response['items'])
    df_supplemented = pd.concat([df_supplemented, df_additional])
```

```python
df_comments = pd.json_normalize(response['items'])
df_comments['snippet.topLevelComment.snippet.textDisplay']
```

```python
i = 1
while 'nextPageToken' in response.keys():
    request = youtube.commentThreads().list(
        part = part,
        videoId = videoId,
        maxResults = amount,
        pageToken = response['nextPageToken'])
    response = request.execute()
    
    add = pd.json_normalize(response['items'])
    df_comments = pd.concat([df_comments, add])
    
    print(f'iteration {i}')
    i += 1
```

```python
df_comments = df_comments.drop_duplicates('id')
df_comments
```

```python
df_comments = pd.DataFrame()
df_video_new = video_df.head(50)
start = time.time()
warnings = 0

for video in df_video_new['id']:
    try:
        request = youtube.commentThreads().list(
            part = part,
            videoId = video,
            maxResults = amount)
        
        response = request.execute()
        
        add = pd.json_normalize(response['items'])
        df_comments = pd.concat([df_comments, add])
        
        print(f'Getting comments for {video}')
        
        i = 1
        while 'nextPageToken' in response.keys():
            request = youtube.commentThreads().list(
                part = part,
                videoId = video,
                maxResults = amount,
                pageToken = response['nextPageToken'])
            response = request.execute()

            add = pd.json_normalize(response['items'])
            df_comments = pd.concat([df_comments, add])

            print(f'iteration {i}')
            i += 1
            
    except: 
        print(f"WARNING error {video}")
        warnings += 1

end = time.time()
print(f'Worked for {int(end - start)} secs')
print(f'Errors = {warnings}')
```

```python
df_comments.index = range (1, len(df_comments)+1)
comments_replies_index = df_comments[df_comments['snippet.totalReplyCount'] > 0].index
```

```python
all_rep_comments = pd.DataFrame()

n = 1
start = time.time()
for i in comments_replies_index:
    df_comments.loc[i, 'Difference'] = df_comments['snippet.totalReplyCount'][i] - len(pd.json_normalize(df_comments['replies.comments'][i]))
    df_add = pd.json_normalize(df_comments['replies.comments'][i])
    all_rep_comments = pd.concat([all_rep_comments, df_add])
    
    print (f'iteration № {n}')
    n += 1 

end = time.time()
print(f'All iterations: {n} for {end - start} secs')
all_rep_comments
```
